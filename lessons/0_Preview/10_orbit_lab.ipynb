# Astronomical Data

````python
from skyfield.api import load

planets = load('de421.bsp')  # JPL ephemeris file
earth, mars, venus, moon = planets['earth'], planets['mars'], planets['venus'], planets['moon']

# Example: get Earth's orbital velocity around the Sun today
ts = load.timescale()
t = ts.now()
earth_at_t = earth.at(t)
velocity = earth_at_t.velocity.km_per_s

print("Earth orbital velocity (km/s):", velocity)
````
